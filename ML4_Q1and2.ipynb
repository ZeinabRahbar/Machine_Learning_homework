```python
import keras
from keras import models 
from keras import layers 
import matplotlib.pyplot as plt
from keras.datasets import mnist
```

```python
# load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# preprocess the data
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

# 1_a

```python
# define the one layer model

model_1 = models.Sequential()
model_1.add(layers.Dense(10, activation='softmax'))


# compile the model
model_1.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# train the model
history_1 = model_1.fit(train_images, train_labels, epochs=20, batch_size=128, validation_split=0.2)
```

```python
# define the two layer model

model_2 = models.Sequential()
model_2.add(layers.Dense(16, activation='relu'))
model_2.add(layers.Dense(10, activation='softmax'))

# compile the model
model_2.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# train the model
history_2 = model_2.fit(train_images, train_labels, epochs=20, batch_size=128, validation_split=0.2)
```

```python
# define the three layer model

model_3 = models.Sequential()
model_3.add(layers.Dense(16, activation='relu'))
model_3.add(layers.Dense(16, activation='relu'))
model_3.add(layers.Dense(10, activation='softmax'))

# compile the model_3
model_3.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# train the model
history_3 = model_3.fit(train_images, train_labels, epochs=20, batch_size=128, validation_split=0.2)
```

```python
# Plot training loss for all three models over epochs
plt.plot(history_1.history['loss'])
plt.plot(history_2.history['loss'])
plt.plot(history_3.history['loss'])
plt.title('Training Loss Comparison')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['One-Layer Model', 'Two-Layer Model', 'Three-Layer Model'], loc='upper right')
plt.show()

# Plot validation accuracy for all three models over epochs
plt.plot(history_1.history['val_accuracy'])
plt.plot(history_2.history['val_accuracy'])
plt.plot(history_3.history['val_accuracy'])
plt.title('Validation Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['One-Layer Model', 'Two-Layer Model', 'Three-Layer Model'], loc='lower right')
plt.show()
```
